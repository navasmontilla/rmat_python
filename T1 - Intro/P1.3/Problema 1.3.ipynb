**Inicialización**

Lo primero que debemos hacer es cargar las librerías necesarias de Python. Para este problema, en particular, será de gran utilidad la librería *sympy* que permite realizar cálculo simbólico (como haríamos con Máxima, Maple o Mathemática).

````python
from sympy import *             # Librería para trabajo simbólico
import math                     # Librería para utilizar símbolos matemáticos como el número pi, que se escribe como math.pi
import numpy as np              # Librería para poder trabajar con matrices y vectores
import matplotlib.pyplot as plt # Librería para poder dibujar gráficas
````

**Cálculo de las reacciones**

Para calcular las reacciones aplicaremos las ecuaciones de equilibrio estático global. Formularemos un sistema de 3 ecuaciones $ΣFx=0$, $ΣFy=0$ y $ΣMz=0 $ con 3 incógnitas, $R_{A,x}$, $R_{A,y}$ y $R_{B,y}$ y lo resolveremos utilizando la función *linsolve*.

````python
Rax,Ray,Rby=symbols('Rax Ray Rby') #variables simbólicas: reacciones en apoyos
Nx1,Nx2,Vy1,Vy2,Mz1,Mz2,Nx1_1,Vy1_1,Mz1_1=symbols('Nx1,Nx2,Vy1,Vy2,Mz1,Mz2,Nx1_1,Vy1_1,Mz1_1') #variables simbólicas: esfuerzos 
L,Lt,NL,PL,HM,Hm,AT,g1,g2,a,b,P,sigma,q1,q2=symbols('L,Lt,NL,PL,HM,Hm,AT,g1,g2,a,b,P,sigma,q1,q2')
x,xc1,xc2=symbols('x,xc1,xc2')

L=3.0 #longitud de cada tramo 
Lt=2*L #longitud de la viga
NL=341 #número de ladrillos
PL=6.6*10 #peso de cada ladrillo
HM= 6 #altura máxima
Hm=HM/2 
x1=1
x2=2

g1=HM*x/L #g1 ecuación altura ladrillos primer tramo
#g2 ecuación altura ladrillos segundo tramo --> cáclulo:
sol,=linsolve([ 
                HM-a*L-b,  #g2(x=L) 
                Hm-a*Lt-b, #g2(x=2*L)
             ], 
                (a,b)  #Es la lista de incógnitas a resolver
            ) 
a=sol[0] 
b=sol[1] 
g2=a*x+b
print("g2 = (",g2, ") m")
print("g1 = ",g1, "m")
AT=integrate(g1, (x, 0, L))+integrate(g2, (x, L, Lt)) # Área total
print("Área total=",AT, "m2")
P=NL*PL #peso total, 
print("Peso total =",P , "N")
sigma=P/AT 
print("Densidad=",sigma, "N/m2")
q1=g1*sigma
q2=g2*sigma
#equilibrio estático, cálculo de las reacciones
sol,=linsolve([ 
                Rax,  
                Ray+Rby-P,
                -integrate(x*q1, (x, 0, L))-integrate(x*q2, (x, L, Lt))+Lt*Rby,
             ], 
                (Rax,Ray,Rby)  #Es la lista de incógnitas a resolver
            ) 
#print (sol)  #La solución está contenida en el vector "sol"
Rax=sol[0] 
Ray=sol[1] 
Rby=sol[2] 
print("Rax=",Rax, "N")
print("Ray=",Ray, "N")
print("Rby=",Rby, "N")
````

**Cálculo de los esfuerzos **


````python
print("-- Cálculo de esfuerzos en tramo 1 --")
#xc1=1 #para resolver en x=1 o x=2 descomentar y poner valor correcto
sol,=linsolve([ 
                Rax+Nx1,  
                Ray-integrate(q1, (x, 0, xc1))+Vy1,
                -integrate(x*q1, (x, 0, xc1))+Vy1*xc1+Mz1,
             ], 
                (Nx1,Vy1,Mz1)  #Es la lista de incógnitas a resolver
            ) 
#print (sol)  #La solución está contenida en el vector "sol"

Nx1=sol[0] 
Vy1=sol[1] 
Mz1=sol[2] 

print("Nx=", Nx1)
print("Vy=", Vy1)
print("Mz=", Mz1)

````

````python
print("-- Cálculo de esfuerzos en tramo 2 --")
#xc2=4 #para resolver en x=4 descomentar y poner valor correcto
sol,=linsolve([ 
                Rax+Nx2,  
                Ray-integrate(q1, (x, 0, L))-integrate(q2, (x, L, xc2))+Vy2,
                -integrate(x*q1, (x, 0, L))-integrate(x*q2, (x, L, xc2))+Vy2*xc2+Mz2,
             ], 
                (Nx2,Vy2,Mz2)  #Es la lista de incógnitas a resolver
            ) 
#print (sol)  #La solución está contenida en el vector "sol"

Nx2=sol[0] 
Vy2=sol[1] 
Mz2=sol[2] 

print("Nx=", Nx2)
print("Vy=", Vy2)
print("Mz=", Mz2)

````
